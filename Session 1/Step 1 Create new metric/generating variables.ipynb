```python
# we want to categorize funding rounds > 100 million as 1 and others as 0

# create a new list to store the new variable
funding_round_category = []

# transform the column Money Raised Currency (in USD) into a list
money_raised = df['Money Raised Currency (in USD)'].tolist()

# loop through the list and categorize the funding rounds
for i in range(len(money_raised)):
    if money_raised[i] > 100000000:
        funding_round_category.append(1)
    else:
        funding_round_category.append(0)
```

```python
# we need to take into account that there are some missing values in the column Money Raised Currency (in USD)

# create a new list to store the new variable
funding_round_category = []
# loop through the list and categorize the funding rounds
for i in range(len(money_raised)):
    if pd.isna(money_raised[i]):
        funding_round_category.append(None)
    else:
        if money_raised[i] > 100000000:
            funding_round_category.append(1)
        else:
            funding_round_category.append(0)

# add the new variable to the dataframe
df['Funding Round Category 100M'] = funding_round_category
```

```python
# check number of values in the new variable for each category
df['Funding Round Category 100M'].value_counts(dropna=False)
```

```python
# provide overview of values on Funding Type column
df['Funding Type'].value_counts(dropna=False)
```

```python
# create binary variable where score is 1 when company engage in post-seed round and 0 otherwise

# create a new list to store the new variable
post_seed_round = []

# transform the column Funding Type into a list
funding_type = df['Funding Type'].tolist()

# loop through the list and categorize the funding rounds

for i in range(len(funding_type)):
    if funding_type[i] == 'Series A' or funding_type[i] == 'Series B' or funding_type[i] == 'Series C' or funding_type[i] == 'Series D' or funding_type[i] == 'Series E' or funding_type[i] == 'Series F' or funding_type[i] == 'Series G' or funding_type[i] == 'Series H' or funding_type[i] == 'Series I' or funding_type[i] == 'Series J' or funding_type[i] == 'Series K' or funding_type[i] == 'Series L' or funding_type[i] == 'Series M' or funding_type[i] == 'Series N' or funding_type[i] == 'Series O' or funding_type[i] == 'Series P' or funding_type[i] == 'Series Q' or funding_type[i] == 'Series R' or funding_type[i] == 'Series S' or funding_type[i] == 'Series T' or funding_type[i] == 'Series U' or funding_type[i] == 'Series V' or funding_type[i] == 'Series W' or funding_type[i] == 'Series X' or funding_type[i] == 'Series Y' or funding_type[i] == 'Series Z':
        post_seed_round.append(1)
    else:
        post_seed_round.append(0)

```

```python
# provide a summary of the new variable
df['Post Seed Round'] = post_seed_round
df['Post Seed Round'].value_counts(dropna=False)
```
