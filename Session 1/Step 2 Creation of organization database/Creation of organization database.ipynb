```python
# import the necessary packages
import pandas as pd

# quick environment check
print('pandas version:', pd.__version__)
try:
    import openpyxl
    print('openpyxl version:', openpyxl.__version__)
except Exception as e:
    print('openpyxl not available:', e)
```

```python
# read the excel file in python as a dataframe

startdata = pd.read_excel('Transactions with whu connection.xlsx', sheet_name='Sheet1')
```

```python
# click on variables and on startdata to see the data
# check the size of the dataframe

startdata.shape
```

```python
# create a list of column names

startdata.columns
```

```python
# create a list of columns to drop

dropcols = ['Transaction Name', 'Transaction Name URL', 'Funding Type', 'Money Raised', 'Money Raised Currency',
       'Money Raised Currency (in USD)', 'Announced Date', 'Investor Names',
       'Funding Stage', 'Number of Investors']
```

```python
# create a new dataframw without columns to drop

data = startdata.drop(dropcols, axis=1)

# A DataFrame object has two axes: “axis 0” and “axis 1”. “axis 0” represents rows and “axis 1” represents columns

# it is wise to give a new name to the new dataframe
```

```python
# double check size of the new dataframe

data.shape
```

```python
# remove duplicates based on column Organization Name URL

dataclean = data.drop_duplicates(subset='Organization Name URL', keep='first').reset_index(drop=True)
```

```python
dataclean.shape
```

```python
# save dataframe as excel file

dataclean.to_excel('organizations.xlsx')
```

```python
# create variable that is 1 when Total Funding Amount (in USD) > 100 million

big_funded = list()

totalfunding = dataclean['Total Funding Amount (in USD)'].to_list()

for i in range(len(totalfunding)):
    if pd.isna(totalfunding[i]):
        big_funded.append(None)
    else:        
        if totalfunding[i] > 100000000:
            big_funded.append(1)
        else:
            big_funded.append(0)

# add variable to dataframe

dataclean['Big Funded'] = big_funded
```

```python
# show values for new variable

dataclean['Big Funded'].value_counts(dropna=False)
```
