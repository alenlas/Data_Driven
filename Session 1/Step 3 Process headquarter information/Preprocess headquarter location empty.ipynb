```python
import pandas as pd
```

```python
# read organization database

data = pd.read_excel(r'organizations.xlsx')
```

```python
# check data
data.head()
```

```python
# turn the column 'Organization Location' into a list

readlist = list(data['Organization Location'])
```

```python
#check the length of the list

print(len(readlist))
```

```python
#check the first 5 observations

readlist[0:5]
```

```python
# Lets take the first observation

locationfirst = readlist[0]
```

```python
locationfirst

# this is a string with 4 elements
```

```python
# split this string into a list consisting of 4 elements

locationfirstsplit = locationfirst.split(',')
```

```python
# show the result

locationfirstsplit

# result is a list with 4 elements; PROBLEM: element 2,3,4 start with a space
```

```python
# redefine split to make sure that elements do not start with a space

locationfirstsplitsecond = locationfirst.split(', ')
locationfirstsplitsecond
```

```python
# to select the city, we need to select the first element of the list

cityfirst = locationfirstsplitsecond[0]
cityfirst

# IMPORTANT: index of list starts with 0
```

```python
# to select the region, we need to select the second element of the list

regionfirst = locationfirstsplitsecond[1]
regionfirst
```

```python
# we need to do this for all observations

# create empty lists

citylist = list()
regionlist = list()

# loop over all observations

totalobservations = len(readlist)

for i in range(0, totalobservations):
    locationlist = readlist[i].split(', ')
    city = locationlist[0]
    region = locationlist[1]
    citylist.append(city)
    regionlist.append(region)
```

```python
# transform the list into a column of the dataframe

data['City'] = pd.DataFrame(citylist)
data['Region'] = pd.DataFrame(regionlist)
```

```python
# provide overview of number of observations per region

data['Region'].value_counts()
```

```python
# provide overview of number of observations per city

data['City'].value_counts()   #Munich, München and Muenchen
```

```python
# drop city column

data.drop(columns=['City'], inplace=True)
```

```python
#replace character in a string

oldstring = 'München'

newstring = oldstring.replace('ü', 'ue')

newstring
```

```python
# replace umlaut in citylist

newcitylist = list()

for i in range(len(citylist)):
    newcity = citylist[i].replace('ü', 'ue').replace('ö', 'oe').replace('ä', 'ae').replace('Ä', 'Ae').replace('Ö', 'Oe').replace('Ü', 'Ue')
    newcitylist.append(newcity)

newcitylist[0:10]
```

```python
# replace Munich into Muenchen

for i in range(len(newcitylist)):
    if newcitylist[i] == 'Munich':  # make sure that in the if statemen you use == and you end with :
        newcitylist[i] = 'Muenchen'

newcitylist[0:10]
```

```python
# create new City column

data['City'] = pd.DataFrame(newcitylist)
```

```python
# check values for City column
data['City'].value_counts() 
```
