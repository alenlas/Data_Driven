```python
# import pandas as pd

import pandas as pd
```

```python
# Read in the data

data = pd.read_excel(r'Transactions with whu connection.xlsx')
```

```python
# turn announced date into year month day

daylist = []
monthlist = []
yearlist = []

for i in range(len(data['Announced Date'])):
    datestring = str(data['Announced Date'][i])
    if '-' in datestring:
        datestringsplit = datestring.split('-')
        daylist.append(float(datestringsplit[2]))   # float statement transforms string to float
        monthlist.append(float(datestringsplit[1]))
        yearlist.append(float(datestringsplit[0]))
    else:
        datestringsplit = datestring.split('/')
        daylist.append(float(datestringsplit[2]))
        monthlist.append(float(datestringsplit[1]))
        yearlist.append(float(datestringsplit[0]))
```

```python
daylist[0:5]
```

```python
data['Announced Day'] = daylist
data['Announced Month'] = monthlist
data['Announced Year'] = yearlist
```

```python
data.to_excel(r'transactions preprocessed.xlsx')
```
