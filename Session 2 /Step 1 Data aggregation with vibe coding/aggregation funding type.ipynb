```python
import pandas as pd
```

```python
# read excel

df = pd.read_excel('transactions preprocessed.xlsx')
```

```python
#check columns

df.columns
```

```python
# select all observations with funding type 'Seed'
df_seed = df[df['Funding Type'] == 'Seed']

# create list with all uniqe values for Announced Year
announced_years = df_seed['Announced Year'].unique()

#sort the list

announced_years.sort()
```

```python
announced_years
```

```python
# create empty list to store the number of transactions per year

seed_transactions_list = []

# loop through the list of announced years and count the number of transactions for each year, then append the count to the list

for year in announced_years:
    seed_transactions_year = df_seed[df_seed['Announced Year'] == year].shape[0]
    seed_transactions_list.append(seed_transactions_year)
```

```python
# check the list

seed_transactions_list
```

```python
# create unique list of funding types
funding_types = df['Funding Type'].unique()

```

```python
# create empty dictionary to store the number of transactions per year for each funding type
funding_type_transactions = {}

# loop through the list of funding types, then for each funding type loop through the list of announced years and count the number of transactions for each year, then append the count to the list and add the list to the dictionary with the funding type as key
for funding_type in funding_types:
    df_funding_type = df[df['Funding Type'] == funding_type]
    transactions_list = []
    for year in announced_years:
        transactions_year = df_funding_type[df_funding_type['Announced Year'] == year].shape[0]
        transactions_list.append(transactions_year)
    funding_type_transactions[funding_type] = transactions_list
```

```python
# check the dictionary

funding_type_transactions
```

```python
#transform dictionary to dataframe
funding_type_transactions_df = pd.DataFrame(funding_type_transactions, index=announced_years)
```

```python
# check the dataframe

funding_type_transactions_df
```

```python
# transform name column Seed into column Number Seed Transactions
funding_type_transactions_df = funding_type_transactions_df.rename(columns={'Seed': 'Seed Transactions'})

# rename every column with name of funding type + ' Transactions'
funding_type_transactions_df = funding_type_transactions_df.rename(columns={funding_type: funding_type + ' Transactions' for funding_type in funding_types})
```

```python
# check the dataframe

funding_type_transactions_df
```

```python
# create empty dictionary to store the number of transactions per year for each funding type for WHU companies
funding_type_transactions_whu = {}

# loop through the list of funding types, then for each funding type loop through the list of announced years and count the number of transactions for each year for WHU companies, then append the count to the list and add the list to the dictionary with the funding type as key
for funding_type in funding_types:
    df_funding_type = df[df['Funding Type'] == funding_type]
    transactions_list = []
    for year in announced_years:
        transactions_year = df_funding_type[(df_funding_type['Announced Year'] == year) & (df_funding_type['WHU Company'] == 1)].shape[0]
        transactions_list.append(transactions_year)
    funding_type_transactions_whu[funding_type] = transactions_list
```

```python
# transform dictionary to dataframe

funding_type_transactions_whu_df = pd.DataFrame(funding_type_transactions_whu, index=announced_years)
```

```python
# rename every column with name of funding type + ' Transactions'
funding_type_transactions_whu_df = funding_type_transactions_whu_df.rename(columns={funding_type: funding_type + ' WHU Transactions' for funding_type in funding_types})
```

```python
# check the dataframe

funding_type_transactions_whu_df
```

```python
# merge both dataframes on index
funding_type_transactions_merged = pd.merge(funding_type_transactions_df, funding_type_transactions_whu_df, left_index=True, right_index=True)
```

```python
# check the merged dataframe

funding_type_transactions_merged.head()
```

```python
# export to excel

funding_type_transactions_merged.to_excel('funding_type_transactions_merged.xlsx')
```

```python
# aggregate data for funding type by Money Raised Currency (in USD)

# start with seed only

seed_money_list = []

for year in announced_years:
    seed_money_year = df_seed[df_seed['Announced Year'] == year]['Money Raised Currency (in USD)'].sum()
    seed_money_list.append(seed_money_year)

seed_money_list

```

```python
# for every funding type, create list with money raised per year

funding_type_money = {}
for funding_type in funding_types:
    df_funding_type = df[df['Funding Type'] == funding_type]
    money_list = []
    for year in announced_years:
        money_year = df_funding_type[df_funding_type['Announced Year'] == year]['Money Raised Currency (in USD)'].sum()
        money_list.append(money_year)
    funding_type_money[funding_type] = money_list

# create dataframe from dictionary

funding_type_money_df = pd.DataFrame(funding_type_money, index=announced_years)

# rename every column with name of funding type + ' Money Raised (in USD)'
funding_type_money_df = funding_type_money_df.rename(columns={funding_type: funding_type + ' Money Raised (in USD)' for funding_type in funding_types})

funding_type_money_df
```

```python
# we need to address missing values. If no transactions exist value should be NaN instead of 0
funding_type_money = {}
for funding_type in funding_types:
    df_funding_type = df[df['Funding Type'] == funding_type]
    money_list = []
    for year in announced_years:
        if df_funding_type[df_funding_type['Announced Year'] == year].shape[0] == 0:
            money_year = None

        else:
            money_year = df_funding_type[df_funding_type['Announced Year'] == year]['Money Raised Currency (in USD)'].sum()
        money_list.append(money_year)
    funding_type_money[funding_type] = money_list

# create dataframe from dictionary

funding_type_money_df = pd.DataFrame(funding_type_money, index=announced_years)

# rename every column with name of funding type + ' Money Raised (in USD)'
funding_type_money_df = funding_type_money_df.rename(columns={funding_type: funding_type + ' Money Raised (in USD)' for funding_type in funding_types})

funding_type_money_df
```

```python
# additional issues with missing values. Transactions that did not disclose money raised should not be considered

# we need to address missing values. If no transactions exist value should be NaN instead of 0
funding_type_money = {}
for funding_type in funding_types:
    df_funding_type = df[df['Funding Type'] == funding_type]
    money_list = []
    for year in announced_years:
        if df_funding_type[df_funding_type['Announced Year'] == year].shape[0] == 0:
            money_year = None

        else:
            cleaned_df = df_funding_type[(df_funding_type['Announced Year'] == year) & (df_funding_type['Money Raised Currency (in USD)'].notna())]
            if cleaned_df.shape[0] == 0:
                money_year = None
            else:
                money_year = cleaned_df['Money Raised Currency (in USD)'].sum()
        money_list.append(money_year)
    funding_type_money[funding_type] = money_list

# create dataframe from dictionary

funding_type_money_df = pd.DataFrame(funding_type_money, index=announced_years)

# rename every column with name of funding type + ' Money Raised (in USD)'
funding_type_money_df = funding_type_money_df.rename(columns={funding_type: funding_type + ' Money Raised (in USD)' for funding_type in funding_types})

funding_type_money_df
```

```python
# for every funding type, create list with money raised from whu companies per year
funding_type_whu_money = {}
for funding_type in funding_types:
    df_funding_type = df[df['Funding Type'] == funding_type]
    df_whu = df_funding_type[df_funding_type['WHU Company'] == 1]
    money_list = []
    for year in announced_years:
        if df_whu[df_whu['Announced Year'] == year].shape[0] == 0:
            money_year = None
        else:
            cleaned_df = df_whu[(df_whu['Announced Year'] == year) & (df_whu['Money Raised Currency (in USD)'].notna())]
            if cleaned_df.shape[0] == 0:
                money_year = None
            else:
                money_year = cleaned_df['Money Raised Currency (in USD)'].sum()
        money_list.append(money_year)
    funding_type_whu_money[funding_type] = money_list

# create dataframe from dictionary
funding_type_whu_money_df = pd.DataFrame(funding_type_whu_money, index=announced_years)

# rename every column with name of funding type + ' WHU Money Raised (in USD)'
funding_type_whu_money_df = funding_type_whu_money_df.rename(columns={funding_type: funding_type + ' WHU Money Raised (in USD)' for funding_type in funding_types})

funding_type_whu_money_df
```

```python
# merge both dataframes on index

funding_type_money_merged = pd.merge(funding_type_money_df, funding_type_whu_money_df, left_index=True, right_index=True, how='outer')  
funding_type_money_merged.head()
```

```python
# export to excel
funding_type_money_merged.to_excel('funding_type_money_merged.xlsx')
```
